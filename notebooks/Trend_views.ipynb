```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel("/content/stock_details_5_years.xlsx", sheet_name='AAPL_stock_vol')
df1.head()
```

```python
df2 = pd.read_excel("/content/Real_GDP_in_Excel.xlsx", sheet_name='financeAndInsurance_RealGDP')
df2.head()
```

```python
df3 = pd.read_excel("/content/Major indices in the Nigeria Capital Markets.xlsx", sheet_name='NGX_ASI' )
df3.head()
```

```python
df4 = pd.read_excel("/content/forex_multiindex.xlsx", sheet_name='USD_NGN')
df4.head()
```

```python
df5 = pd.read_excel("/content/CMO-Historical-Data-Monthly.xlsx", sheet_name='Brent')
df5.head()
```

```python
df1['date'] = pd.to_datetime(df1['date'])
df1 = df1.set_index('date')

df2['date_period'] = pd.to_datetime(df2['date_period'])
df2 = df2.set_index('date_period')

df3['date'] = pd.to_datetime(df3['date'])
df3 = df3.set_index('date')

df4['date'] = pd.to_datetime(df4['date'])
df4 = df4.set_index('date')

df5['date'] = pd.to_datetime(df5['date'])
df5 = df5.set_index('date')
```

```python
for dataframe in [df1, df2, df3, df4, df5]:
  display(dataframe.head())
  print('\n\n')
```

```python
df1['avr_monthly_megavol'].plot(title="Monthly Average Mega-Volume of Trades for AAPL", figsize=(10,5))
plt.xlabel("Date")
plt.ylabel("Monthly Average")
plt.show()
```

```python
df2['financeAndInsurance'].plot(title="Monthly Average Real GDP for Finance & Insurance Sector", figsize=(10,5))
plt.xlabel("Date")
plt.ylabel("Monthly Average")
plt.show()
```

```python
df3['avr_monthly_kiloprice'].plot(title="Monthly Average Price (k) for NGX ASI Index", figsize=(10,5))
plt.xlabel("Date")
plt.ylabel("Monthly Average")
plt.show()
```

```python
df4[['avr_open', 'avr_close']].plot(title="Monthly Average Open and Close Price for USD/NGN Index", figsize=(10,5))
plt.xlabel("Date")
plt.ylabel("Monthly Average")
plt.show()
```

```python
df5['brent_crude_oil ($/bbl)'].plot(title="Monthly Average Global Brent Crude Oil Price ($/bbl)", figsize=(10,5))
plt.xlabel("Date")
plt.ylabel("Monthly Average")
plt.show()
```
